# **Project Summary -**
The Netflix Movies and TV Shows Clustering project is intended to conduct an in-depth exploratory data analysis (EDA) of the content library of Netflix to reveal patterns, trends, and insights that can facilitate improved business decision-making. The dataset employed includes data such as the title, type (Movie or TV Show), director, cast, country, release year, duration, genre (listed_in), and date added to the platform. The analysis began with data cleaning, where missing values were identified and handled appropriately. For instance, missing values in the ‘date_added’ column were managed using type conversion, and string fields like ‘country’ and ‘listed_in’ were cleaned for consistency.

Various visualizations were created using libraries such as Matplotlib and Seaborn to help interpret the data visually. Some key takeaways from the data are that movies take up the majority of Netflix's content library, representing a higher percentage than television shows. A bar chart plotting the number of Movies vs. TV Shows demonstrated this quite effectively. The year-by-year temporal analysis of content added revealed a steep increase post-2015, which indicates Netflix's explosive content growth and international expansion during that period. Genre-wise, the most popular categories are Dramas, Comedies, and Documentaries, reflecting viewer interest in storytelling and informative programming.

Another significant discovery was the leading content-producing countries on Netflix, as the United States, India, and the United Kingdom were found to be most frequent contributors. This indicates how much the platform depends on regional and English-language content. Second, the study examined content ratings, and from that, they found that the most prevalent rating is TV-MA (Mature Audience) and suggests there is an attempt at targeting a mature audience in their strategy.

The analysis also investigated the length of films, revealing that the majority of films are between 80–120 minutes, and looked at the most common directors working on the platform. Each visualization was chosen to best represent the type of data—bar charts for comparing categorical data, line charts for trends in time series data, and histograms for distribution analysis.

In summary, this project assisted in visualizing and interpreting the nature of Netflix's content library. It provided insights that can benefit Netflix in content planning, audience segmentation, and regional strategy formulation. Knowledge of trends in genres, countries, release patterns, and content types allows Netflix to improve user satisfaction and remain competitive in the expanding digital streaming market.

# **Problem Statement**-
The problem addressed in this project revolves around the difficulty users face in discovering relevant content within Netflix's ever-expanding library of movies and TV shows. As the platform continues to grow, traditional filtering methods based on genre tags or search queries often fall short in capturing the deeper similarities between titles. This leads to a less personalized and sometimes overwhelming user experience. To tackle this issue, the project focuses on applying unsupervised machine learning techniques—specifically clustering algorithms—to group similar content based on key features such as genre, cast, country, release year, and duration. By identifying natural groupings within the data, the project aims to uncover hidden patterns that can enhance content discovery, improve recommendation systems, and assist Netflix in better organizing and managing its vast collection. This approach not only benefits users by making content navigation more intuitive but also supports strategic decisions for content curation and personalization on the platform.

#### **Define Your Business Objective?**
The main business goal of this project is to increase user satisfaction and engagement on the Netflix site by enhancing how content is categorized and recommended. By using clustering methods to classify similar movies and TV shows according to common attributes—genre, cast, country, release year, and duration—Netflix can provide more personalized and accurate recommendations to its users. This results in greater content discoverability, increased user watch times, and lower churn. Furthermore, the knowledge obtained through clustering can be used to make strategic decisions in content acquisition, marketing, and user segmentation. Overall, the aim is to leverage data-driven methods to improve understanding of content relationships and optimize the user experience, ultimately driving competitiveness and growth in the streaming market.

## ***Data Vizualization***

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

##**Chart-1** Count of Movies vs TV Shows



```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', color='blue')
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

The Count of Movies vs TV Shows chart is employed for examining the general distribution of content types on Netflix. It indicates whether the platform contains more movies or more TV shows, which is vital for obtaining a high-level perspective on the dataset prior to proceeding with in-depth analysis.



##### 2. What are the insight(s) found from the chart?
As can be seen from the chart, Netflix supports a much higher number of movies than TV programs. This would imply that Netflix gives more emphasis to providing more extensive movie cataloging, possibly in order to satisfy a greater diversity of user preferences and supply more one-off viewing opportunities rather than serialized programs.

##**Chart-2**  Top 10 Countries by Content Count

```python
plt.figure(figsize=(10,5))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, color='green')
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
To determine what genres predominate Netflix's content catalog and what types of content occur most often.
##### 2. What are the insight(s) found from the chart?
International Movies are the most prevalent genre on Netflix.Others include Dramas, Comedies, and International TV Shows.Genres such as Romantic Movies, Children & Family Movies, and Independent Movies have comparatively fewer titles. This indicates that Netflix focuses on globally popular and story-driven genres.

##**Chart-3** Titles Added Per Year

```python
# Convert date_added to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(10,5))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='teal')
plt.title("Titles Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
In order to examine how the number of titles added to Netflix has shifted through the years and identify patterns of growth.
##### 2. What are the insight(s) found from the chart?
There was a huge spike in titles being added from 2016 to 2019.The highest was in 2019, indicating Netflix's most aggressive content push.Few titles were added prior to 2015.There was a decline in 2020, perhaps because the COVID-19 pandemic disrupted production timelines.



##**Chart-4** Top 10 Most Common Genres

```python
plt.figure(figsize=(10,5))
genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)
sns.barplot(x=genres.values, y=genres.index, color='yellow')
plt.title("Top 10 Most Common Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

##**Chart-5** Content Added Each Year by Type

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='year_added', color='orange')
plt.title("Content Added Every Year by Type")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
This chart illustrates the trend of content growth on Netflix throughout the years and the shift in the balance between Movies and TV Shows.
##### 2. What is/are the insight(s) found from the chart?
Netflix has grown its content base much more since approximately 2015
Movies tend to outnumber TV Shows per year.A spike in content addition might indicate strategic growth or licensing trends of content.

##**Chart-6** Content Added by Month

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='month_added', color='pink')
plt.title("Content Added by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

##**Chart-7** Top 10 Directors with Most Titles

```python
top_directors = df['director'].dropna().value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='crest')
plt.title("Top 10 Directors with Most Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
In order to discover which directors appear most commonly on Netflix—significant for viewers, analysts, or partnership assessments.

##### 2. What is/are the insight(s) found from the chart?
There are directors with heavy presence, perhaps owing to partnerships or region-specific content.It also shows how significant regional or high-volume creators are in defining the offerings of the platform.



##**Chart-8** Distribution of Movie Durations

```python
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_num'] = movie_df['duration'].str.extract('(\\d+)').astype(float)

plt.figure(figsize=(10, 5))
sns.histplot(movie_df['duration_num'].dropna(),color='orange')
plt.title("Distribution of Movie Durations (in minutes)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
A histogram is well-suited to illustrate how film lengths are distributed on the platform—helpful for creators or analysts.
##### 2. What is/are the insight(s) found from the chart?
The majority of films are 80–120 minutes, consistent with typical feature film lengths.A couple of outliers indicate extremely short or long films exist, perhaps documentaries or specials.



##**Chart-9** Distribution of TV Show Seasons

```python
tv_df = df[df['type'] == 'TV Show'].copy()
tv_df['seasons_num'] = tv_df['duration'].str.extract('(\\d+)').astype(float)

plt.figure(figsize=(10, 5))
sns.countplot(x=tv_df['seasons_num'], color ='Purple')
plt.title("Distribution of TV Show Seasons")
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")
plt.tight_layout()
plt.show()


```

##### 1. Why did you pick the specific chart?
TV Shows differ significantly in the number of seasons they have; this number provides an idea of how much content to expect from a viewer per show.
##### 2. What is/are the insight(s) found from the chart?
Most shows have 1 or 2 seasons, showing either new shows or Netflix's practice of staying away from long-running shows.Minimal shows have more than 3+ seasons, perhaps due to expense or viewer engagement practices.


##**Chart-10** Top 10 Countries Producing Most Content

```python
top_countries = df['country'].dropna().str.split(', ', expand=True).stack().value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
This horizontal bar chart is selected to display the quantity of titles contributed by the leading countries. Horizontal bars are easier to read for long country names and compare side by side.
##### 2. What is/are the insight(s) found from the chart?
United States is far ahead with the most number of titles contributed. India, United Kingdom, and Canada are also prominent content contributors. This indicates Netflix's robust global production strategy and localization efforts.



##**Chart-11** Content Rating Distribution

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, color='brown')
plt.title("Distribution of Content Ratings")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
This bar graph is great to use to demonstrate each rating's frequency (such as TV-MA, PG, etc.) in order to understand Netflix's target audience for content.
##### 2. What is/are the insight(s) found from the chart?
Most titles are rated TV-MA (Mature Audience), which suggests that Netflix aims to reach adult audiences heavilyFewer titles are rated TV-Y or G, demonstrating less focus on children's shows.The range reveals how varied the content is within various age categories.
