```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Dataset
df1=pd.read_csv('/content/titles.csv')
df2=pd.read_csv('/content/credits.csv')
```

```python
#Merging the two data sets into single data set
df3=pd.merge(df1,df2,on='id',how='left')
```

```python
df3
```

```python
df3.info()
```

```python
df3.shape
```

```python
df3.describe()
```

```python
#Finding the missing values
df3.isnull()
```

```python
#Sorting the missing values
df3.isnull().sum().sort_values(ascending=False)
```

```python
df3.dropna(inplace=True)
```

```python
df3.info()
```

```python
# Fill missing values with 0
df3.fillna(0, inplace=True)
```

```python
# Finding null value count
df3.isnull().sum()
```

##**Chart-1** Content Type Distribution

```python
df1['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

##### 1. Why did you pick the specific chart?
When comparing categorical data, bar charts work best. Here, we contrast the number of movies and TV shows.

##### 2. What are the insight(s) found from the chart?
The platform's dominating content may be revealed by the fact that movies appear more frequently than TV shows, or vice versa.



##**Chart-2** Content Released Per Year

```python
df3['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5), color='green')
plt.title('Number of Titles Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid()
plt.show()
```

##**Chart-3** Top 10 Genres

```python
df3['genres'].value_counts().head(10).plot(kind='barh', color='orange')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

##**Chart-4** imdb Distribution

```python
sns.histplot(df3['imdb_score'], bins=20, color='purple')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.show()
```

##### 1. Why did you pick the specific chart?
The distribution of numerical values is displayed by histograms. Here, we apply it to examine trends in IMDb scores.

##### 2. What are the insight(s) found from the chart?
Indicates the distribution of content quality (as measured by ratings); do the majority of shows receive average ratings (5-7) or extremes.






##**Chart-5** Average IMDB score By content Type

```python
df3.groupby('type')['imdb_score'].mean().plot(kind='bar', color='lightcoral')
plt.title('Average IMDb Score by Content Type')
plt.ylabel('Average Score')
plt.show()
```

##### 1. Why did you pick the specific chart?
This displays a comparison of means between categories, making it ideal for determining which type performs better overall.

##### 2. What are the insight(s) found from the chart?
The average rating of TV shows may be marginally higher or lower than that of movies, indicating a perception of quality.



##**Chart-6** Boxplot of IMDb Scores by Type

```python
sns.boxplot(data=df3, x='type', y='imdb_score', palette='Set2')
plt.title('IMDb Scores by Type (Boxplot)')
plt.xlabel('Content Type')
plt.ylabel('IMDb Score')
plt.show()
```

##### 1. Why did you pick the specific chart?
For comprehending distribution, median, and outliers within categories, boxplots are a great tool.

##### 2. What are the insight(s) found from the chart?
You might discover that there are more rating fluctuations for movies or fewer outliers for shows.



##**Chart-7**  Content Count by Age Certification

```python
df3['age_certification'].value_counts().head(10).plot(kind='bar', color='steelblue')
plt.title('Top Age Certifications')
plt.xlabel('Certification')
plt.ylabel('Count')
plt.show()
```

##### 1. Why did you pick the specific chart?
Once more, categorical comparison aids in understanding the platform's certification-based target audience.
##### 2. What are the insight(s) found from the chart?
You might discover that "16+" or "18+" predominate, indicating that the platform values mature content.





##**Chart-8** Top 10 Actors by Appearance

```python
top_actors = df3['name'].str.split(', ').explode().value_counts().head(10)
top_actors.plot(kind='barh', color='mediumpurple')
plt.title('Top 10 Most Frequent Actors')
plt.xlabel('Number of Titles')
plt.ylabel('Actor')
plt.show()
```

##### 1. Why did you pick the specific chart?
Displays how frequently actors appear, which is excellent for casting trends and talent visibility.

##### 2. What are the insight(s) found from the chart?
You can find actors who are featured in several productions or who regularly collaborate with Amazon.





##**Chart-9** Top 10 production countries

```python
top_countries = df3['production_countries'].value_counts().head(10)
top_countries.plot(kind='bar', color='salmon')
plt.title('Top 10 Production countries')
plt.xlabel('countries')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

##### 1. Why did you pick the specific chart?
When comparing values across distinct categories, like production countries in this case, a bar chart works best. It assists in determining which nations contribute the most to the content library of Amazon Prime.


##### 2. What are the insight(s) found from the chart?
According to the chart, the US produces the most content for Amazon Prime, followed by Canada, India, and the UK. With some contributions from around the world, the majority of the content is from the US.






##**Chart-10** Top 10 Genres

```python
top_genres = df3['genres'].value_counts().head(10)
plt.figure(figsize=(6,6))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=60, colors=plt.cm.Pastel1.colors)
plt.title('Top 10 Genres Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

##### 1. Why did you pick the specific chart?
To display the most popular genres' proportionate distribution. Comparing category shares at a glance is made simple by pie charts.

##### 2. What are the insight(s) found from the chart?
Comedy and drama may be the most popular genres, indicating a focus on platform content.Genres are revealed by smaller slices (e.g., Documentary, Horror).Aids in comprehending Amazon Prime's content strategy and viewer preferences.
